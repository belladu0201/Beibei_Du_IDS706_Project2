### Extract all the words and flatten the N-D list

```python
with open('Kanye_West_org.txt') as f:
    words = []
    lines = f.readlines()
    for line in lines:
        words.append(line.split())
flattened = [sublist for val in words for sublist in val]
#print(flattened)
```

### Get a list of English stopwords

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stops = set(stopwords.words('english'))
#print(stops)
```

### Store the Non-stop words into a new list called `ans`

```python
ans = []
for i in flattened:
    if i not in stops:
        ans.append(i)
len(ans) == len(flattened)
```

### Find out the top 50 words in Kanye West lyrics

```python
# find the most appeared words in Kanye West's lyrics
from collections import Counter
Counter = Counter(ans)
most_occur = Counter.most_common(50)
print(most_occur)
```

## Create a Word Cloud for Reference


```python
# Reference: https://www.datacamp.com/tutorial/wordcloud-python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(" ".join(ans))
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
